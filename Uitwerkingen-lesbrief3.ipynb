```python
import pandas as pd

dfGames = pd.read_csv("games.csv")

dfGames.columns

kolommen = ["Name", "Platform", "EU_Sales"]

dfGamesKlein = dfGames[kolommen]

dfGamesKleinSorted = dfGamesKlein.sort_values(by=["EU_Sales"], ascending=True)

dfPSGames = dfGamesKleinSorted[dfGamesKleinSorted.Platform == "PS"]
dfPSGames.head(10)
```
